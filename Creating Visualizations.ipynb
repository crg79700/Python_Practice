```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing Ice Cream Ratings into a data frame
df = pd.read_csv(r"C:\Users\colin\OneDrive\Documents\Data Science\Python\Pandas\Ice Cream Ratings.csv")
df = df.set_index("Date")
df
```
