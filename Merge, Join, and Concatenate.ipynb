```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\colin\OneDrive\Documents\Data Science\Python\Pandas\LOTR.csv")

df1
```

```python
df2 = pd.read_csv(r"C:\Users\colin\OneDrive\Documents\Data Science\Python\Pandas\LOTR 2.csv")
df2
```

```python
# Don't specify the merge, it will automatically put in an inner merge, inner joins only match stuff that is the same in both tables
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
# Outer gives all values regardless if they are the same or not.
df1.merge(df2, how = 'outer')
```

```python
# Takes everything from the left data frame, then pulls everything that matches in the second data frame.
df1.merge(df2, how = 'left')
```

```python
# Does the exact opposite of the left join. (Pulls data frame 2 and matches only stuff that matches from data frame 1
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
# Join works best with indexes not with column names that well. In the next cell, I'll show how to do the index
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right', how = 'outer')

df3
```

```python
# Concatenating dataframes puts one data frame on top of the other rather than next to each other (like merge and join)
pd.concat([df1, df2], join = 'outer', axis = 1)
```
