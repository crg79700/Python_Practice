```python
import pandas as pd
```

```python
# One way of setting up the index if you already know what variable you want to have as the index
df = pd.read_csv(r"C:\Users\colin\OneDrive\Documents\Data Science\Python\Pandas\world_population.csv", index_col = 'Country')

df
```

```python
# How you reset the index if you realize that you messed up the indexing
df.reset_index(inplace = True)

df
```

```python
# Another way of setting the index
df.set_index('Country', inplace = True)

df
```

```python
# You can use the Index to filter through your data quite easily
# You can use loc (location) and iloc (integer location) to do this

df.loc['Albania']
```

```python
# iloc looks for the integer position of the item in the data frame (starting with 0 as the first row)
df.iloc[1]
```

```python
# Multi-Indexing
df.reset_index(inplace = True)
df
```

```python
df.set_index(['Continent', 'Country'], inplace = True)

df
```

```python
df.sort_index()
```

```python
#Loc can specify by the multi-index
df.loc['Africa', 'Angola']
```

```python
#iloc does not specify by multi-indexing. It only specifies by the original first index.
df.iloc[1]
```
